```python
import os
import numpy as np
import pandas as pd
from skimage import io, measure
from skimage.io import imsave
from skimage.util import img_as_ubyte
from cellpose import models
from tqdm import tqdm
input_root = '/content/SPLIT500'
output_root = '/content/CellNet_masks'
model = models.CellposeModel(gpu=True, model_type='cyto')
print("✅ Cellpose ")
for label in os.listdir(input_root):
    label_path = os.path.join(input_root, label)
    if not os.path.isdir(label_path):
        continue
    out_class_dir = os.path.join(output_root, label)
    os.makedirs(out_class_dir, exist_ok=True)

    image_files = [
        f for f in os.listdir(label_path)
        if f.lower().endswith(('.jpg', '.png', '.jpeg', '.tif'))
    ]
    all_data = []
    for fname in tqdm(image_files, desc=f"[{label}]"):
        try:
            img_path = os.path.join(label_path, fname)
            img = io.imread(img_path)
            if img.ndim == 2:
                img = np.stack([img] * 3, axis=-1)
            elif img.shape[-1] > 3:
                img = img[:, :, :3]
            masks, flows, styles = model.eval(img, diameter=None, channels=[0, 0])
            fname_base = os.path.splitext(fname)[0]
            out_mask_path = os.path.join(out_class_dir, fname_base + '_mask.png')
            binary_mask = img_as_ubyte(masks > 0)
            imsave(out_mask_path, binary_mask)

            props = measure.regionprops(masks)
            for p in props:
                y, x = p.centroid
                minr, minc, maxr, maxc = p.bbox
                all_data.append({
                    'image_name': fname,
                    'y': y,
                    'x': x,
                    'area': p.area,
                    'min_row': minr,
                    'min_col': minc,
                    'max_row': maxr,
                    'max_col': maxc
                })
        except Exception as e:
            print(f"❌ error {fname} in [{label}]: {e}")
    csv_path = os.path.join(out_class_dir, f"{label}_nuclei.csv")
    pd.DataFrame(all_data).to_csv(csv_path, index=False)
    print(f"✅ nuclei saved: {csv_path}")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
class EncoderProjection(nn.Module):
    def __init__(self, input_channels=1, latent_dim=128, projection_dim=64):
        super().__init__()
        # Encoder:nucleus ROI
        self.encoder = nn.Sequential(
            nn.Conv2d(input_channels, 32, kernel_size=3, stride=1, padding=1),  # 64x64 → 64x64
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1),  # 64x64 → 32x32
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1),  # 32x32 → 16x16
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1),  # 16x16 → 8x8
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1, 1))  # 8x8 → 1x1
        )
        self.flatten = nn.Flatten()
        self.fc_latent = nn.Linear(256, latent_dim)
        # Projection Head
        self.projection_head = nn.Sequential(
            nn.Linear(latent_dim, latent_dim),
            nn.ReLU(),
            nn.Linear(latent_dim, projection_dim)
        )
    def forward(self, x):
        x = self.encoder(x)       # → (B, 256, 1, 1)
        x = self.flatten(x)       # → (B, 256)
        z = self.fc_latent(x)     # → (B, latent_dim)
        z = F.normalize(z, dim=1)
        p = self.projection_head(z)
        p = F.normalize(p, dim=1)
        return z, p
model = EncoderProjection()
dummy = torch.randn(8, 1, 64, 64)  # 8 64
z, p = model(dummy)
print("z.shape:", z.shape)  # → torch.Size([8, 128])
print("p.shape:", p.shape)  # → torch.Size([8, 64])
```

```python
class Decoder(nn.Module):
    def __init__(self, latent_dim=128, output_channels=1, output_size=64):
        super().__init__()
        self.init_size = output_size // 8  # three time upsampling 64
        self.fc = nn.Linear(latent_dim, 256 * self.init_size * self.init_size)
        self.decoder = nn.Sequential(
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Upsample(scale_factor=2),  # 8 → 16
            nn.Conv2d(256, 128, 3, stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Upsample(scale_factor=2),  # 16 → 32
            nn.Conv2d(128, 64, 3, stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Upsample(scale_factor=2),  # 32 → 64
            nn.Conv2d(64, output_channels, 3, stride=1, padding=1),
            nn.Sigmoid()  # because output binary
        )
    def forward(self, z):
        x = self.fc(z)                             # (B, 256 * 8 * 8)
        x = x.view(z.size(0), 256, self.init_size, self.init_size)  # (B, 256, 8, 8)
        out = self.decoder(x)                      # (B, 1, 64, 64)
        return out
decoder = Decoder(latent_dim=128, output_channels=1, output_size=64)
z_dummy = torch.randn(8, 128)
recon = decoder(z_dummy)
print("recon.shape:", recon.shape)  # → torch.Size([8, 1, 64, 64])
```

```python
class CAME(nn.Module):
    def __init__(self, input_channels=1, latent_dim=128, projection_dim=64, output_size=64):
        super().__init__()
        self.encoder_proj = EncoderProjection(
            input_channels=input_channels,
            latent_dim=latent_dim,
            projection_dim=projection_dim
        )
        self.decoder = Decoder(
            latent_dim=latent_dim,
            output_channels=input_channels,
            output_size=output_size
        )

    def forward(self, x):
        """
        Args:
            x: Input image tensor of shape (B, C, H, W)
        Returns:
            z: latent representation
            p: projection head output (for contrastive loss)
            recon: reconstructed image
        """
        z, p = self.encoder_proj(x)  # → (B, latent), (B, proj)
        recon = self.decoder(z)      # → (B, C, H, W)
        return z, p, recon
model = CAME(input_channels=1, latent_dim=128, projection_dim=64, output_size=64)
dummy_input = torch.randn(8, 1, 64, 64)
z, p, recon = model(dummy_input)

print("z:", z.shape)       # [8, 128]
print("p:", p.shape)       # [8, 64]
print("recon:", recon.shape)  # [8, 1, 64, 64]
```

```python
import torch.nn.functional as F

def reconstruction_loss(x, recon):
    return F.mse_loss(recon, x)
import torch

def nt_xent_loss(p, temperature=0.5):
    """
    SimCLR-style contrastive loss. Assumes input p is normalized.
    Args:
        p: Tensor of shape (2N, dim)
    Returns:
        Scalar loss
    """
    batch_size = p.size(0)
    assert batch_size % 2 == 0, "Batch size must be even for SimCLR."
    # Compute cosine similarity matrix
    sim_matrix = F.cosine_similarity(p.unsqueeze(1), p.unsqueeze(0), dim=2)  # shape: (2N, 2N)
    # Mask to remove self-similarity from numerator
    mask = torch.eye(batch_size, dtype=torch.bool, device=p.device)
    sim_matrix.masked_fill_(mask, -9e15)
    # Labels: i-th sample matches (i + N) if i < N
    labels = torch.cat([torch.arange(batch_size // 2, batch_size), torch.arange(0, batch_size // 2)]).to(p.device)
    logits = sim_matrix / temperature
    loss = F.cross_entropy(logits, labels)
    return loss
```

```python
def total_loss(x, recon, p, alpha=1.0, beta=1.0):
    loss_recon = reconstruction_loss(x, recon)
    loss_contrastive = nt_xent_loss(p)
    return alpha * loss_recon + beta * loss_contrastive, loss_recon.item(), loss_contrastive.item()
```

```python
import os
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image
import numpy as np
from skimage.measure import regionprops, label as sk_label  # ✅ تغییر نام تابع
import torch.optim as optim
from tqdm import tqdm
from skimage.io import imread

class NucleusPairDataset(Dataset):
    def __init__(self, csv_path, image_dir, mask_dir, crop_size=128, transform_rgb=None, transform_mask=None):
        self.data = pd.read_csv(csv_path)
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.crop_size = crop_size
        self.transform_rgb = transform_rgb
        self.transform_mask = transform_mask

    def __len__(self):
        return len(self.data)

    def crop_center(self, img, x, y):
        half = self.crop_size // 2
        left = max(x - half, 0)
        upper = max(y - half, 0)
        right = left + self.crop_size
        lower = upper + self.crop_size
        return img.crop((left, upper, right, lower))

    def _resolve_path(self, base_dir, fname, is_mask=False):
        base_name = os.path.splitext(fname)[0]
        for cls in os.listdir(base_dir):
            class_dir = os.path.join(base_dir, cls)
            for f in os.listdir(class_dir):
                if is_mask:
                    if base_name in f and f.endswith('.png'):
                        return os.path.join(class_dir, f)
                else:
                    if f.startswith(base_name):
                        return os.path.join(class_dir, f)
        raise FileNotFoundError(f"Matching file for {fname} not found in {base_dir}")

    def __getitem__(self, idx):
        row = self.data.iloc[idx]

        fname = row['filename']
        x = int((row['x_min'] + row['x_max']) / 2)
        y = int((row['y_min'] + row['y_max']) / 2)
        class_label = int(row['label'])

        img_path = self._resolve_path(self.image_dir, fname)
        mask_path = self._resolve_path(self.mask_dir, fname, is_mask=True)

        img = Image.open(img_path).convert("RGB")
        mask = Image.open(mask_path).convert("L")

        img_crop = self.crop_center(img, x, y)
        mask_crop = self.crop_center(mask, x, y)

        x1_rgb = self.transform_rgb(img_crop) if self.transform_rgb else img_crop
        x2_rgb = self.transform_rgb(img_crop) if self.transform_rgb else img_crop
        x1_mask = self.transform_mask(mask_crop) if self.transform_mask else mask_crop
        x2_mask = self.transform_mask(mask_crop) if self.transform_mask else mask_crop

        # Morphological feature extraction
        mask_np = np.array(mask)
        mask_labeled = sk_label(mask_np)
        ecc = 0.0
        area = 0.0
        solid = 0.0

        for region in regionprops(mask_labeled):
            cy, cx = region.centroid
            if ((cx - x) ** 2 + (cy - y) ** 2) ** 0.5 < 4:
                ecc = region.eccentricity
                area = region.area
                solid = region.solidity
                break

        area_norm = area / 2000.0  if area > 0 else 0.0
        morph_features = torch.tensor([ecc, area_norm, solid], dtype=torch.float32)

        return x1_rgb, x2_rgb, x1_mask, x2_mask, class_label, morph_features
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms, models
from PIL import Image
import pandas as pd
import numpy as np
from skimage.measure import regionprops, label as sk_label
import os
from tqdm import tqdm
import random

# --------- Dataset Definition ---------

class NucleusPairDataset(Dataset):
    def __init__(self, csv_path, image_dir, mask_dir, crop_size=128, transform=None, augment=True):
        self.data = pd.read_csv(csv_path)
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.crop_size = crop_size
        self.transform = transform
        self.augment = augment

        # Data augmentation transforms
        self.aug_transform = transforms.Compose([
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomVerticalFlip(p=0.5),
            transforms.RandomRotation(10),
            transforms.ColorJitter(brightness=0.2, contrast=0.2),
            transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
        ])

    def __len__(self):
        return len(self.data)

    def crop_center(self, img, x, y):
        half = self.crop_size // 2
        left = max(x - half, 0)
        upper = max(y - half, 0)
        right = left + self.crop_size
        lower = upper + self.crop_size
        return img.crop((left, upper, right, lower))

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        fname = row['filename']
        base_name = os.path.splitext(fname)[0]
        mask_filename = base_name + "_mask.png"

        x = int((row['x_min'] + row['x_max']) / 2)
        y = int((row['y_min'] + row['y_max']) / 2)

        mask_path = None
        # Iterate through subdirectories (classes) within mask_dir to find the mask
        for cls in os.listdir(self.mask_dir):
            path = os.path.join(self.mask_dir, cls, mask_filename)
            if os.path.exists(path):
                mask_path = path
                break

        if mask_path is None:
            raise FileNotFoundError(f"Could not find mask file for {mask_filename} in {self.mask_dir} or its subdirectories.")

        mask = Image.open(mask_path).convert("L") # Open mask in grayscale
        mask_crop = self.crop_center(mask, x, y)

        if self.augment:
            x1 = self.aug_transform(mask_crop)
            x2 = self.aug_transform(mask_crop)
        else:
            x1 = mask_crop
            x2 = mask_crop

        if self.transform:
            x1 = self.transform(x1)
            x2 = self.transform(x2)
        else: # Fallback to ToTensor if no transform is provided
            to_tensor = transforms.ToTensor()
            x1 = to_tensor(x1)
            x2 = to_tensor(x2)

        # Morphological features extraction
        mask_np = np.array(mask_crop)

        if mask_np.sum() == 0:
            ecc, area, solid = 0.0, 0.0, 0.0
        else:
            mask_labeled = sk_label(mask_np > 0)
            ecc, area, solid = 0.0, 0.0, 0.0

            regions = regionprops(mask_labeled)
            if regions:
                main_region = max(regions, key=lambda r: r.area)
                ecc = main_region.eccentricity
                area = main_region.area / (self.crop_size * self.crop_size)
                solid = main_region.solidity
            else:
                ecc, area, solid = 0.0, 0.0, 0.0

        morph = torch.tensor([ecc, area, solid], dtype=torch.float32)
        return x1, x2, morph

# --------- Attention Module ---------

class SelfAttention(nn.Module):
    def __init__(self, in_channels):
        super().__init__()
        self.query = nn.Conv2d(in_channels, in_channels//8, 1)
        self.key = nn.Conv2d(in_channels, in_channels//8, 1)
        self.value = nn.Conv2d(in_channels, in_channels, 1)
        self.gamma = nn.Parameter(torch.zeros(1))

    def forward(self, x):
        batch_size, C, H, W = x.size()

        q = self.query(x).view(batch_size, -1, H*W).permute(0, 2, 1)
        k = self.key(x).view(batch_size, -1, H*W)
        v = self.value(x).view(batch_size, -1, H*W)

        attention = F.softmax(torch.bmm(q, k), dim=2)
        out = torch.bmm(v, attention.permute(0, 2, 1))
        out = out.view(batch_size, C, H, W)

        return self.gamma * out + x

# --------- Model Definition ---------

class SwinBackbone(nn.Module):
    def __init__(self, output_dim=512):
        super().__init__()
        self.backbone = models.swin_t(weights="DEFAULT")
        self.backbone.head = nn.Identity()

        self.feature_extractor = nn.Sequential(
            nn.Linear(768, 1024),
            nn.LayerNorm(1024),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(1024, output_dim),
            nn.LayerNorm(output_dim)
        )

    def forward(self, x):
        if x.shape[1] == 1:
            x = x.repeat(1, 3, 1, 1)
        feat = self.backbone(x)
        return self.feature_extractor(feat)

class ResidualBlock(nn.Module):
    def __init__(self, channels):
        super().__init__()
        self.conv1 = nn.Conv2d(channels, channels, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(channels)
        self.conv2 = nn.Conv2d(channels, channels, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(channels)
        self.attention = SelfAttention(channels)

    def forward(self, x):
        residual = x
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        out = self.attention(out)
        out += residual
        return F.relu(out)

class BetterDecoder(nn.Module):
    def __init__(self, latent_dim=512, output_channels=1, output_size=128):
        super().__init__()
        self.init_size = output_size // 8

        self.fc = nn.Sequential(
            nn.Linear(latent_dim, 512 * self.init_size * self.init_size),
            nn.LayerNorm(512 * self.init_size * self.init_size),
            nn.ReLU(),
            nn.Dropout(0.1)
        )

        self.decoder = nn.ModuleList([
            nn.Sequential(
                nn.ConvTranspose2d(512, 256, 4, stride=2, padding=1),
                nn.BatchNorm2d(256),
                nn.ReLU(),
                ResidualBlock(256),
                nn.Dropout(0.1)
            ),
            nn.Sequential(
                nn.ConvTranspose2d(256, 128, 4, stride=2, padding=1),
                nn.BatchNorm2d(128),
                nn.ReLU(),
                ResidualBlock(128),
                nn.Dropout(0.1)
            ),
            nn.Sequential(
                nn.ConvTranspose2d(128, 64, 4, stride=2, padding=1),
                nn.BatchNorm2d(64),
                nn.ReLU(),
                ResidualBlock(64)
            ),
            nn.Sequential(
                nn.Conv2d(64, 32, kernel_size=3, padding=1),
                nn.BatchNorm2d(32),
                nn.ReLU(),
                ResidualBlock(32),
                nn.Conv2d(32, output_channels, kernel_size=3, padding=1),
                nn.Sigmoid()
            )
        ])

    def forward(self, z):
        x = self.fc(z)
        x = x.view(z.size(0), 512, self.init_size, self.init_size)

        for block in self.decoder:
            x = block(x)

        return x

class MorphSimCLRModel(nn.Module):
    def __init__(self, feat_dim=512, proj_dim=128, recon_size=128):
        super().__init__()
        self.encoder = SwinBackbone(output_dim=feat_dim)

        self.projector = nn.Sequential(
            nn.Linear(feat_dim, feat_dim),
            nn.LayerNorm(feat_dim),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(feat_dim, proj_dim)
        )

        self.morph_head = nn.Sequential(
            nn.Linear(feat_dim, 256),
            nn.LayerNorm(256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 128),
            nn.LayerNorm(128),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(128, 3)
        )

        self.decoder = BetterDecoder(latent_dim=feat_dim, output_channels=1, output_size=recon_size)

    def forward(self, x):
        z = self.encoder(x)

        p = F.normalize(self.projector(z), dim=1)
        morph = self.morph_head(z)
        recon = self.decoder(z)

        return p, morph, recon

# --------- Training Loop Helpers ---------

# Dice Loss for segmentation
def dice_loss(pred, target, smooth=1e-5):
    pred = pred.contiguous()
    target = target.contiguous()
    intersection = (pred * target).sum(dim=(2,3))
    union = pred.sum(dim=(2,3)) + target.sum(dim=(2,3))
    dice = (2. * intersection + smooth) / (union + smooth)
    return 1 - dice.mean()

# Combined Reconstruction Loss (BCE + Dice)
def reconstruction_loss(pred, target, alpha=0.5):
    bce = F.binary_cross_entropy(pred, target)
    dice = dice_loss(pred, target)
    return alpha * bce + (1 - alpha) * dice

# NT-Xent Loss for contrastive learning (ADDED/MODIFIED)
def nt_xent_loss(p, temperature=0.5):
    """
    SimCLR-style contrastive loss. Assumes input p is normalized.
    Args:
        p: Tensor of shape (2N, dim)
    Returns:
        Scalar loss
    """
    batch_size = p.size(0)
    assert batch_size % 2 == 0, "Batch size must be even for SimCLR."
    sim_matrix = F.cosine_similarity(p.unsqueeze(1), p.unsqueeze(0), dim=2)
    # Mask out self-similarities and positive pairs to avoid trivial solutions
    # The current `labels` construction implicitly handles positive pairs.
    # We only need to mask out self-similarity (diagonal)
    mask = torch.eye(batch_size, dtype=torch.bool, device=p.device)
    # Set diagonal to a very small number to avoid division by zero and large logits
    sim_matrix.masked_fill_(mask, -9e15)

    # Create labels for positive pairs
    # E.g., for batch_size=4, labels would be [2, 3, 0, 1]
    # This means (p[0], p[2]) are positive, (p[1], p[3]) are positive
    labels = torch.cat([
        torch.arange(batch_size // 2, batch_size),
        torch.arange(0, batch_size // 2)
    ]).to(p.device)

    logits = sim_matrix / temperature
    loss = F.cross_entropy(logits, labels)
    return loss


def train_one_epoch(model, loader, optimizer, device):
    model.train()
    total_loss = 0
    total_contrast = 0
    total_morph = 0
    total_recon = 0

    # Weights for the multi-task loss
    w_contrast = 0.3
    w_morph = 0.4
    w_recon = 0.3

    pbar = tqdm(loader, desc="Training")
    for x1_batch, x2_batch, morph_batch in pbar:
        # Move inputs to device
        x1_batch = x1_batch.to(device)
        x2_batch = x2_batch.to(device)
        morph_batch = morph_batch.to(device)

        # Concatenate x1 and x2 to process them as a single batch for efficiency
        x_combined = torch.cat([x1_batch, x2_batch], dim=0)
        # Repeat morphological labels for the combined batch
        morph_combined = morph_batch.repeat(2, 1)

        # Forward pass for all tasks
        p_combined, morph_pred_combined, recon_combined = model(x_combined)

        # Split outputs back for individual loss calculations (not needed for NT-Xent on p_combined)
        # p1, p2 = p_combined.chunk(2, dim=0) # No longer explicitly needed for NT-Xent directly
        recon1, recon2 = recon_combined.chunk(2, dim=0)

        # Contrastive Loss (SimCLR part) - NOW USING NT-Xent Loss
        loss_contrast = nt_xent_loss(p_combined)

        # Morphological Prediction Loss
        loss_morph = F.mse_loss(morph_pred_combined, morph_combined)

        # Reconstruction Loss (for both original and augmented images)
        loss_recon = (reconstruction_loss(recon1, x1_batch) + reconstruction_loss(recon2, x2_batch)) / 2.0

        # Combine all losses
        loss = (w_contrast * loss_contrast +
                w_morph * loss_morph +
                w_recon * loss_recon)

        # Optimization step
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Update and display progress
        total_loss += loss.item()
        total_contrast += loss_contrast.item()
        total_morph += loss_morph.item()
        total_recon += (reconstruction_loss(recon1, x1_batch).item() + reconstruction_loss(recon2, x2_batch).item()) / 2.0

        pbar.set_postfix({
            "Total": f"{total_loss / (pbar.n + 1):.4f}",
            "Recon": f"{total_recon / (pbar.n + 1):.4f}",
            "Contrast": f"{total_contrast / (pbar.n + 1):.4f}",
            "Morph": f"{total_morph / (pbar.n + 1):.4f}"
        })

def validate_one_epoch(model, val_loader, device):
    model.eval()
    total_loss = 0
    total_batches = 0

    w_contrast = 0.3
    w_morph = 0.4
    w_recon = 0.3

    with torch.no_grad():
        for x1, x2, morph in val_loader:
            x1, x2, morph = x1.to(device), x2.to(device), morph.to(device)

            x_combined = torch.cat([x1, x2], dim=0)
            morph_combined = morph.repeat(2, 1)

            p_combined, morph_pred_combined, recon_combined = model(x_combined)
            # p1, p2 = p_combined.chunk(2, dim=0) # Not needed for NT-Xent
            recon1, recon2 = recon_combined.chunk(2, dim=0)

            # Contrastive Loss - NOW USING NT-Xent Loss
            loss_contrast = nt_xent_loss(p_combined)
            loss_morph = F.mse_loss(morph_pred_combined, morph_combined)
            loss_recon = (reconstruction_loss(recon1, x1) + reconstruction_loss(recon2, x2)) / 2

            loss = w_contrast * loss_contrast + w_morph * loss_morph + w_recon * loss_recon

            total_loss += loss.item()
            total_batches += 1

    return total_loss / total_batches

def run_training(csv_path, image_dir, mask_dir, num_epochs=1, batch_size=16, crop_size=32):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    transform = transforms.Compose([
        transforms.Resize((crop_size, crop_size)),
        transforms.ToTensor()
    ])

    full_dataset = NucleusPairDataset(csv_path, image_dir, mask_dir, crop_size=crop_size, transform=transform, augment=True)

    total_size = len(full_dataset)
    val_size = int(0.2 * total_size)
    train_size = total_size - val_size

    train_dataset, val_dataset = random_split(full_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, drop_last=True,
                              num_workers=os.cpu_count(), pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, drop_last=False,
                            num_workers=os.cpu_count(), pin_memory=True)

    model = MorphSimCLRModel(recon_size=crop_size).to(device)
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=0.01)
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(
        optimizer, mode='min', factor=0.5, patience=2, verbose=True
    )

    os.makedirs("checkpoints", exist_ok=True)
    print(f"Starting training with batch_size={batch_size}, crop_size={crop_size}")

    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")

        model.train()
        train_one_epoch(model, train_loader, optimizer, device)

        val_loss = validate_one_epoch(model, val_loader, device)
        tqdm.write(f"Validation Loss: {val_loss:.4f}")
        scheduler.step(val_loss)

        model_path = f"checkpoints/model_epoch_{epoch+1}.pth"
        torch.save(model.state_dict(), model_path)
        print(f"✅ Model saved: {model_path}")

if __name__ == "__main__":
    run_training("/content/detection_annotations.csv", "/content/drive/MyDrive/SPLIT_IMAGES", "/content/drive/MyDrive/CellNet_masks", num_epochs=1, batch_size=16, crop_size=48)
```
